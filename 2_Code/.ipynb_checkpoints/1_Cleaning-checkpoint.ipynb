```python
import pandas as pd
import numpy as np
import os
import glob
from tqdm import tqdm
pd.options.display.float_format = '{:,.4f}'.format
pd.set_option('display.max_columns', None)
```

```python
def clean_id_number(column):
    """
    Convert Indonesian formatted strings (e.g., '1.500,50') to numeric floats.
    
    Args:
        column (pd.Series): Raw string/object column from IDX Excel.
    Returns:
        pd.Series: Cleaned float64 numeric series.
    """
    # Standardize separators: Remove dots (thousands) and fix commas (decimals)
    clean_col = (column.astype(str)
                 .str.strip()
                 .str.replace('.', '', regex=False)
                 .str.replace(',', '.', regex=False)
                 .str.replace(r'[^\d.]', '', regex=True))
    
    return pd.to_numeric(clean_col, errors='coerce')
```

```python
def extract_date_from_filename(filename):
    """
    Extract date from IDX filename and convert to a datetime object.
    
    Expected format: '...-YYYYMMDD.xlsx'
    """
    try:
        # Isolate the YYYYMMDD string between the hyphen and the file extension
        date_part = filename.split('-')[-1].split('.')[0]
        
        return pd.to_datetime(date_part, format='%Y%m%d')
    
    except Exception as e:
        # Log failure to help identify corrupted or misnamed historical files
        print(f"Warning: Could not extract date from {filename}. Error: {e}")
        return None
```

```python
def calculate_investment_metrics(df):
    """
    Calculate return (Y), foreign intensity (X1), and volume (X2) metrics.
    
    Converts key columns to float to ensure precision and prevent rounding errors.
    """
    # Define columns to be processed as floats
    cols = ['Penutupan', 'Sebelumnya', 'Foreign Buy', 'Foreign Sell', 'Volume']
    
    # Casting and calculating in a clean, readable flow
    df = df.astype({col: float for col in cols if col in df.columns})

    # Metric Calculations
    df['Y_Price_Return'] = (df['Penutupan'] - df['Sebelumnya']) / df['Sebelumnya'].replace(0, np.nan)
    df['X1_Foreign_Intensity'] = (df['Foreign Buy'] - df['Foreign Sell']) / (df['Volume'] + 1)
    df['X2_Volume'] = df['Volume']
    
    return df
```

```python
"""
Portfolio Universe Filtering
Focus: Mid-Cap growth and value hunters.
Strategy: Filters Indonesian stocks listed on the 'Utama' and 'Pengembangan' boards.
"""

# 1. Initialize Stock Universe Metadata
# Using a glob pattern to stay resilient against changing export dates in filenames
UNIVERSE_PATTERN = '../1_Raw_Data/Daftar_Saham/Daftar Saham *.xlsx'
matching_files = glob.glob(UNIVERSE_PATTERN)

# Data Integrity Check: Ensure at least one metadata file exists before proceeding
if not matching_files:
    raise FileNotFoundError(f"Critical Error: No metadata file found matching pattern: {UNIVERSE_PATTERN}")

# Select the most recent file by sorting alphabetically (YYYYMMDD format ensures chronological order)
latest_metadata_path = sorted(matching_files)[-1]
saham_master = pd.read_excel(latest_metadata_path)

print(f"Successfully loaded universe from: {os.path.basename(latest_metadata_path)}")
```

```python
# 2. Board Selection (Universe Gatekeeper)
# We prioritize 'Utama' and 'Pengembangan' boards to ensure we analyze established 
# companies, effectively filtering out high-risk speculative or transitionary tiers.
saham_pilihan = ['Utama', 'Pengembangan']
saham_seleksi = saham_master[saham_master['Papan Pencatatan'].isin(saham_pilihan)].copy()

# 3. Dimensionality Reduction / Feature Selection
# Pruning the dataset to include only the attributes required for 
# Valuation (Market Cap) and Identification (Ticker/Company Name).
saham_seleksi = saham_seleksi[['Kode', 'Nama Perusahaan', 'Saham', 'Papan Pencatatan']]

# 4. Integrity Verification
# Visualizing the head of the DataFrame and schema info to ensure 
# the 'Gatekeeper' logic has executed correctly before the main loop.
print(f"\n{'='*30} GATEKEEPER {'='*30}\n")
print(saham_seleksi.head())

print(f"\n{'-' * 30} DATA SCHEMA {'-' * 30}")
print(saham_seleksi.info())
```

```python
# 5. Numerical Data Sanitization
# Applying the custom 'clean_id_number' utility to standardize the 'Saham' column.
# This ensures total shares outstanding are represented as numeric types for calculation.
saham_seleksi['Saham'] = clean_id_number(saham_seleksi['Saham'])

# Verify the data type transformation to prevent 'Object' type arithmetic errors
print(f"Verified Dtype for Saham: {saham_seleksi['Saham'].dtype}")
```

```python
# 6. File Discovery & Path Validation
# Identify all daily trade summaries within the source directory using glob patterns.
trade_folder = '../1_Raw_Data/Ringkasan_Perdagangan_Saham'

if not os.path.exists(trade_folder):
    raise OSError(f"CRITICAL ERROR: Directory not found at '{trade_folder}'. Verify relative pathing.")

# Retrieve and sort files to ensure chronological order based on YYYYMMDD naming convention.
all_trade_files = sorted(glob.glob(os.path.join(trade_folder, "Ringkasan Saham-*.xlsx")))

if not all_trade_files:
    raise FileNotFoundError(f"No valid trade files detected in {trade_folder}.")

# Isolate the most recent trade report for the 'Latest' snapshot analysis.
latest_file_path = all_trade_files[-1]
latest_filename = os.path.basename(latest_file_path)

print(f"Total files found: {len(all_trade_files)}")
print(f"Latest report identified: {latest_filename}")
```

```python
# 7. Data Ingestion & Structural Verification
# Loading the identified Excel file into a pandas DataFrame for processing.
df_latest = pd.read_excel(latest_file_path)

# Verifying dataset architecture and Dtypes to ensure compatibility with downstream metrics.
print(f"\n{'='*20} DATASET ARCHITECTURE {'='*20}\n")
print(df_latest.info())
print(f"\n{'='*60}")
```

```python
# 8. Relational Database Integration
# Merging daily trade data with the 'Gatekeeper' metadata to enrich the dataset 
# with Share Counts and Listing Board information. We use an 'inner' join 
# to ensure we only process tickers present in our curated universe.
saham_merge = pd.merge(
    df_latest, 
    saham_seleksi[['Kode', 'Saham', 'Papan Pencatatan']], 
    left_on='Kode Saham', 
    right_on='Kode', 
    how='inner'
)

# Redundancy Cleanup: Removing duplicate joining keys to maintain a clean schema.
saham_merge = saham_merge.drop(columns=['Kode'])

# 8.1 Post-Merge Integrity Audit
# Displaying formatted snapshots to verify that shares and prices are aligned.
print(f"{'='*20} MERGE VERIFICATION {'='*20}\n")
print(saham_merge[['Kode Saham', 'Penutupan', 'Saham', 'Papan Pencatatan']]
      .assign(Penutupan=lambda x: x['Penutupan'].astype(float),
              Saham=lambda x: x['Saham'].astype(float))
      .head())
```

```python
# 9. Valuation Engine: Market Capitalization Calculation
# We derive the Market Cap by multiplying the closing price by the total shares 
# outstanding. This metric is essential for our Mid-Cap segmentation strategy.
saham_merge['Market Cap'] = saham_merge['Penutupan'] * saham_merge['Saham']

# Data Cleaning: Removing any tickers with missing valuation data to ensure 
# the integrity of our universe filtering.
saham_merge = saham_merge.dropna(subset=['Market Cap'])

# 9.1 Valuation Verification
# Displaying the raw components and the resulting Market Cap with thousands 
# separators for human-readable auditing.
print(f"{'='*20} MARKET CAP VERIFICATION {'='*20}\n")
print(saham_merge[['Kode Saham', 'Penutupan', 'Saham', 'Market Cap']]
      .astype({'Penutupan': float, 'Saham': float, 'Market Cap': float})
      .head())
```

```python
# 10. Strategic Universe Segmentation (Mid-Cap Filtering)
# We isolate the Mid-Cap segment (2T - 20T IDR) to align with our 
# 'Value and Growth' mandate. This filter excludes high-volatility 
# Small-Caps and over-saturated Big-Caps.
batas_bawah = 2_000_000_000_000 
batas_atas = 20_000_000_000_000

# Applying a Boolean mask to generate a standalone Mid-Cap dataframe
saham_mid_cap = saham_merge[
    (saham_merge['Market Cap'] >= batas_bawah) & 
    (saham_merge['Market Cap'] <= batas_atas)
].copy()

# 10.1 Segmentation Analytics & Display
# Reporting the filtering efficiency and displaying top-weighted Mid-Caps 
# using formatted floating-point values for clarity.
print(f"Universe filtered from {len(saham_merge)} to {len(saham_mid_cap)} tickers.")
print(f"Mid-Cap filter successful.")

# Casting to float only within the print statement to trigger comma separators
print(saham_mid_cap[['Kode Saham', 'Market Cap']]
      .sort_values('Market Cap', ascending=False)
      .astype({'Market Cap': float}))
```

```python
# 11. Quantitative Logic Validation (The "Truth Test")
# Before executing the bulk processing loop, we validate the metric engine 
# on a sample subset to ensure floating-point precision and catch any 
# potential 'zero-rounding' or 'division by zero' errors.

# 11.1 Metric Application
# Applying vectorized calculations for Price Return (Y) and Foreign Intensity (X1)
saham_test = calculate_investment_metrics(saham_mid_cap.copy())

# 11.2 Visual Audit
# Verifying that indicators are generating non-zero, high-precision decimals
print(f"{'='*20} TESTING CALCULATION {'='*20}\n")
print(saham_test[['Kode Saham', 'Penutupan', 'Sebelumnya', 'Y_Price_Return', 'X1_Foreign_Intensity']]
    .head(10))

# 11.3 Statistical Sanity Check
# Confirming the mean values are captured in 6-decimal precision to 
# validate the 'Aggressive-Strategic' sensitivity of the model.
print(f"\nMean of Price Return: {saham_test['Y_Price_Return'].mean():.6f}")
print(f"Mean of Foreign Intensity: {saham_test['X1_Foreign_Intensity'].mean():.6f}")
```

```python
# 12. Batch Processing Engine (Historical Data Reconstruction)
# Iteratively processing daily trade files to build a longitudinal dataset
# for the selected Mid-Cap universe.
all_processed_data = []
target_tickers = saham_mid_cap['Kode Saham'].unique()
errors = []

print(f"Initiating batch processing for {len(all_trade_files)} historical files...")

for file in all_trade_files:
    filename = os.path.basename(file)
    try:
        # 12.1 Data Ingestion & Date Extraction
        df_raw = pd.read_excel(file)
        current_date = extract_date_from_filename(filename)
        
        if current_date is None:
            raise ValueError(f"Metadata Error: Date extraction failed for {filename}")

        # 12.2 Dynamic Universe Filtering
        # Restricting the dataset to the predefined 'Gatekeeper' universe (2T - 20T)
        df_filtered = df_raw[df_raw['Kode Saham'].isin(target_tickers)].copy()
        
        # Operational Safeguard: Skip empty snapshots without breaking the loop
        if df_filtered.empty:
            print(f"Notice: No target tickers found in {filename}. Skipping.")
            continue

        # 12.3 Quantitative Processing
        # Appending time-series stamps and calculating growth/value metrics
        df_filtered['Tanggal_Data'] = current_date
        df_ready = calculate_investment_metrics(df_filtered)
        
        all_processed_data.append(df_ready)

    except Exception as e:
        # Resilience Logic: Log anomalies to 'errors' while maintaining loop execution
        error_msg = f"Critical Failure in {filename}: {str(e)}"
        errors.append(error_msg)
        print(error_msg)

# 13. Execution Audit
# Final validation of the batch process integrity.
if errors:
    print(f"\nProcessing completed with {len(errors)} error(s). Review 'errors' list for audit.")
else:
    print("\nBatch processing successful. All historical snapshots synchronized.")
```

```python
# 14. Data Consolidation (Historical Synthesis)
# Flattening the list of daily snapshots into a unified Master Table.
# 'ignore_index=True' is used to ensure a clean, continuous index for the final dataset.
df_master_history = pd.concat(all_processed_data, ignore_index=True)

# 14.1 Spatiotemporal Alignment
# Sorting by Date and Ticker is essential for time-series analysis 
# and tracking the 'Aggressive-Strategic' progression of specific holdings.
# We also move 'Tanggal_Data' to the first position.
df_master_history = df_master_history.sort_values(by=['Tanggal_Data', 'Kode Saham'])
cols = ['Tanggal_Data'] + [col for col in df_master_history.columns if col != 'Tanggal_Data']
df_master_history = df_master_history[cols]

# 15. Final Quantitative Audit
# Validating the consolidated universe size and performing a statistical 
# sanity check on core investment metrics (Y, X1, X2).
print(f"{'='*20} MASTER DATA ARCHITECTURE {'='*20}")
print(f"Final Universe Size: {df_master_history['Kode Saham'].nunique()} Tickers")
print(f"Total Data Points: {len(df_master_history)} rows")

# Reviewing distribution and outliers for Price Return and Foreign Flow
print(f"\n{'='*20} DESCRIPTIVE STATISTICS {'='*20}")
print(df_master_history[['Y_Price_Return', 'X1_Foreign_Intensity', 'X2_Volume']].describe())

# Displaying the synchronized final state
print(f"\n{'='*20} MASTER HEAD VIEW {'='*20}")
print(df_master_history.head())
```

```python
# 15. Automated Filename Generation
# Extracting the start and end dates from the sorted file list to ensure 
# the filename accurately reflects the data duration.
start_date = os.path.basename(all_trade_files[0]).split('-')[1].split('.')[0]
end_date = os.path.basename(all_trade_files[-1]).split('-')[1].split('.')[0]

output_dir = '../2_Processed_Data'
output_filename = f'Master_Universe_{start_date}_{end_date}.csv'
output_path = os.path.join(output_dir, output_filename)

# 15.1 Directory Persistence & Export with Safeguard
try:
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
        print(f"System Update: Created directory '{output_dir}'")

    # Attempt to save the master file
    df_master_history.to_csv(output_path, index=False)
    print(f"Dynamic Export Successful: {output_filename}")

except PermissionError:
    print(f"CRITICAL ERROR: Could not save file at {output_path}.")
    print("REASON: The file is likely open in Excel. Please close it and try again.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
