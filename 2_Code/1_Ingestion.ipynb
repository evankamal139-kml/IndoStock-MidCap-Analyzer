```python
import pandas as pd
import os
import glob
import config

importlib.reload(config)
```

```python
# 1. Load all saham files then make a list.

all_files = glob.glob(os.path.join(config.RAW_DATA_PATH, "Ringkasan Saham-*.xlsx"))
total_files = len(all_files)
all_dataframes = []

for i, file in enumerate(all_files, start=1):
    try:
        df = pd.read_excel(file)
        all_dataframes.append(df)
        print(f"[{i}/{total_files}] Successfully read: {os.path.basename(file)}")
    except FileNotFoundError:
        print(f"Error: The File {file} was not found.")
    except Exception as e:
        print(f"Error on file {i}: {e}")
```

```python
# 2. Combine and output the ingested files.

if all_dataframes:
    master_df = pd.concat(all_dataframes, ignore_index=True)
    output_path = os.path.join(config.INGESTED_DATA_PATH, "Master_Raw_Data.csv")
    master_df.to_csv(output_path, index=False)

    print("-" * 30)
    print(f"SUCCESS! Combined {len(all_dataframes)} files.")
    print(f"Total Rows: {len(master_df)}")
    print(f"Saved to: {output_path}")
else:
    print("No data was collected. Check your file paths.")
```
